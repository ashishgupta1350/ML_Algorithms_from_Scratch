```python
import pandas as pd
import numpy as np
from sklearn import model_selection as ms
from sklearn.linear_model import LinearRegression as LR
# we are finding a boundary for the boston data set using pandas and numpy | and Sklearn
```

```python
from sklearn import datasets

data=datasets.load_boston()
```

```python
#now we have the data, split it into x and y values 
x=data.data
y=data.target
df=pd.DataFrame(x)
df.columns=data.feature_names
df.describe()
```

```python
alg1=LR()
X_train,X_test,Y_train,Y_test=ms.train_test_split(x,y,random_state=0)
alg1.fit(X_train,Y_train)
ypred=alg1.predict(X_test)

alg1.score(X_test,Y_test)
```

```python
df['age_age']=df.AGE*df.AGE
alg2=LR()
x1=df.values
y1=y
X_train1,X_test1,Y_train1,Y_test1=ms.train_test_split(x1,y1,random_state=0)
alg2.fit(X_train1,Y_train1)
ypred1=alg2.predict(X_test1)

alg2.score(X_test1,Y_test1)

```
