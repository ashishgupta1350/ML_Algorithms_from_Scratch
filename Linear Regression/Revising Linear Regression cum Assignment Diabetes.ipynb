```python
import numpy as np
from sklearn import model_selection as ms
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('data_diabetes.csv',delimiter=',')
print data
```

```python

x,y=data[:,0:10],data[:,10]

X_train,X_test,Y_train,Y_test=ms.train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression as LR
alg1=LR() # it creates alg1 as an object of the class LR

alg1.fit(X_train,Y_train)

'''m=alg1.coef_
c=alg1.intercept_

'''
ypred=alg1.predict(X_test)

plt.scatter(Y_test,ypred)
plt.xlabel('Y_test')
plt.ylabel('Y_pred')
l=np.linspace(40,350,1000)
plt.plot(l,l,color='black')
plt.grid()
plt.show()
```

```python
#we need to train the algorithm for all the data sets, it will be quiet accurate
from sklearn.linear_model import LinearRegression as LR
alg1=LR() # it creates alg1 as an object of the class LR

alg1.fit(x,y)
ypred=alg1.predict(X_test)

plt.scatter(Y_test,ypred)
plt.xlabel('Y_test')
plt.ylabel('Y_pred')
l=np.linspace(40,350,1000)
plt.plot(l,l,color='black')
plt.grid()
plt.show()
```

```python
data2test=np.loadtxt('test.csv',delimiter=',')
#data2test has all the xvalues and now we just need to predict these values and make this upto 5 decimal places and store this in a file

ypred_submit=alg1.predict(data2test)
help(np.savetxt)
```
